## Juego del Ahorcado
El funcionamiento es que vaya mostrando por pantalla ```_ _ _ _ _``` las rayas bajas correspondientes al número de letras que tiene la palabra a adivinar. El usuario deberá insertar una letra y, si esa letra existe, se mostrará o mostrarán las letras acertadas ```S _ S _ _```. 

Si falla, se mostrará el número de errores que lleva y, si lleva 6 errores, entonces se mostrará "Fin del juego".
Encapsúlalo en una función.

````python
palabra = input("Palabra para el ahorcado")
número_vidas = 6

ahorcado = list(palabra.upper())
print(ahorcado)

funcion_juego = lambda lista: ["_" for _ in lista]

print(" ".join(funcion_juego(ahorcado)))

progreso = list(funcion_juego(ahorcado))
````

````python
while True:
    if "_" not in progreso:
        print(" ".join(progreso))
        print("¡Has ganado!")
        break
    
    letra = input("Introduce una letra").upper()

    if letra in ahorcado:
        for x, y in enumerate(ahorcado):
            if y == letra:
                progreso[x] = letra
        print(" ".join(progreso))
    else:
        número_vidas -= 1
        if número_vidas <= 0:
            print("Se te han acabado las vidas ")
            break
        print(" ".join(progreso))
        print("Has perdido una vida, te quedan" + " " + str(número_vidas))
````
