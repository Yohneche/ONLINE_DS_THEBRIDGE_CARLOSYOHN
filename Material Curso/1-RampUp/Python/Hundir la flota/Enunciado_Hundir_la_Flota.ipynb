```python
import random

# ===============================
#   CREACI√ìN DE TABLEROS VAC√çOS
# ===============================

tablero_jugador = []
tablero_maquina = []

# Tablero del jugador (10x10)
for i in range(10):
    linea = []
    for j in range(10):
        linea.append(" ")
    tablero_jugador.append(linea)

# Tablero de la m√°quina (10x10)
for i in range(10):
    linea = []
    for j in range(10):
        linea.append(" ")
    tablero_maquina.append(linea)

# ===============================
#   TABLEROS DE DISPAROS
# ===============================

disparos_jugador = []
disparos_maquina = []

for _ in range(10):
    disparos_jugador.append([" "] * 10)
    disparos_maquina.append([" "] * 10)


# ===============================
#   DISPARO ALEATORIO M√ÅQUINA
# ===============================

def disparo_maquina():
    while True:
        x = random.randint(0, 9)
        y = random.randint(0, 9)

        # La m√°quina no repite disparos
        if disparos_maquina[y][x] == " ":
            return x, y


# ===============================
#   COLOCAR BARCOS (AUTOM√ÅTICO)
# ===============================

def colocar_barco(tablero, eslora):
    while True:
        orientacion = random.choice(["H", "V"])  # Horizontal o Vertical

        fila = random.randint(0, 9)
        col = random.randint(0, 9)

        # ---------------------
        # HORIZONTAL
        # ---------------------
        if orientacion == "H":
            if col + eslora > 10:   # no cabe
                continue

            # comprobar si est√° libre
            libre = True
            for i in range(eslora):
                if tablero[fila][col + i] != " ":
                    libre = False
                    break

            if not libre:
                continue

            # colocar barco
            for i in range(eslora):
                tablero[fila][col + i] = "B"
            break

        # ---------------------
        # VERTICAL
        # ---------------------
        else:
            if fila + eslora > 10:
                continue

            libre = True
            for i in range(eslora):
                if tablero[fila + i][col] != " ":
                    libre = False
                    break

            if not libre:
                continue

            # colocar barco
            for i in range(eslora):
                tablero[fila + i][col] = "B"
            break


def colocar_barcos_automatico(tablero):
    # 4 barcos de eslora 1
    for _ in range(4):
        colocar_barco(tablero, 1)

    # 3 barcos de eslora 2
    for _ in range(3):
        colocar_barco(tablero, 2)

    # 2 barcos de eslora 3
    for _ in range(2):
        colocar_barco(tablero, 3)

    # 1 barco de eslora 4
    colocar_barco(tablero, 4)


# ===============================
#   COLOCAR BARCOS EN AMBOS
# ===============================

colocar_barcos_automatico(tablero_jugador)
colocar_barcos_automatico(tablero_maquina)


# ===============================
#   DEBUG OPCIONAL
# ===============================

print("TABLERO DE LA M√ÅQUINA (DEBUG):")
for fila in tablero_maquina:
    print(" ".join(fila))

def imprimir_tablero(titulo, tablero):
    print(f"\n{titulo}")
    print("    " + " ".join(str(i+1).rjust(2) for i in range(10)))
    print("   " + "-" * 30)

    for idx, fila in enumerate(tablero):
        linea = " ".join(fila[i].rjust(2) for i in range(10))
        print(str(idx+1).rjust(2) + " | " + linea)
    print()
```
