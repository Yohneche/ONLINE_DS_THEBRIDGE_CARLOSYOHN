## Ejercicio 11
No queda muy claro cuánto de dispar, de dispersa, es la medida de IBU. Mediante estadística descriptiva (una o dos métricas, sin gráficas), razona la dispersión de la variable IBU.

```python
media = beers["ibu"].mean()
std = beers["ibu"].std()
rango = beers["ibu"].max() - beers["ibu"].min()

media, std, rango
```

La media es 42.7 y el máximo y el mínimo están muy alejados, la dispersión es muy elevada y la media aumenta o decrece mucho.

## Ejercicio 12
¿Existe alguna relación lineal entre las variables numéricas del dataset? Intenta pintar este razonamiento mediante un gráfico

```python
relacion_lineal = beers[["abv", "ibu", "id", "brewery_id", "ounces"]].corr()
relacion_lineal
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
sns.heatmap(relacion_lineal, annot=True, cmap="coolwarm")
plt.show()
```
